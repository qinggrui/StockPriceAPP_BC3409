```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

hazard_type = [ 'Blizzard']

#stormwx1, national weather service, 

# Using TwitterSearchScraper to scrape data and append tweets to list
def TWTRScrapr(hazard_type):
    for j in hazard_type:

    # Created a list to append all tweet attributes(data)
        attributes_container = []

        attributes_container2 = []

        for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:stormwx1 lang:en').get_items()):
            if i>100:
                break
            attributes_container.append([tweet.date, tweet.likeCount, tweet.sourceLabel,  tweet.user.username, tweet.content])
            tweets_df = pd.DataFrame(attributes_container, columns=["Date Created", "Number of Likes", "Source of Tweet", "Username", "Tweets"])

        for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:NWS lang:en').get_items()):
            if i>100:
                break
            attributes_container2.append([tweet.date, tweet.likeCount, tweet.sourceLabel, tweet.user.username, tweet.content])    
# Creating a dataframe from the tweets list above 
            tweets_df2 = pd.DataFrame(attributes_container2, columns=["Date Created", "Number of Likes", "Source of Tweet", "Username", "Tweets"])
            
            merge = pd.concat([tweets_df,tweets_df2], axis=0).sort_values(by=['Date Created'], ascending = False).reset_index(drop=True)

        return merge

TWTRScrapr(hazard_type)

tweet_df = TWTRScrapr(hazard_type)
```

```python
from nltk.tokenize import RegexpTokenizer

regexp = RegexpTokenizer('\w+')

tweet_df['text_token'] = tweet_df['Tweets'].apply(regexp.tokenize)
tweet_df.head(3)
```

```python
import nltk
# nltk.download('punkt')

from textblob import TextBlob
import pandas as pd

# Define a function to check the similarity of the words
def check_similarity(text):
    text_blob = TextBlob(text)
    word = "blizzard"
    return word in text_blob.words

# Apply the function to the text column of the DataFrame
for i in tweet_df[['Tweets']]:
    tweet_df['is_similar'] = tweet_df['Tweets'].apply(check_similarity)

# Filter the DataFrame to obtain only the rows with similar words
blizzard_Tweets = tweet_df[tweet_df['is_similar'] == True] 

blizzard_Tweets
```

```python
# https://www.kaggle.com/code/amar09/sentiment-analysis-on-scrapped-tweets

# for all NLP related operations on text
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import *
from nltk.classify import NaiveBayesClassifier
from wordcloud import WordCloud

from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
from textblob.np_extractors import ConllExtractor

# 1 way
def fetch_sentiment_using_textblob(text):
    analysis = TextBlob(text)
    return 'pos' if analysis.sentiment.polarity >= 0 else 'neg'
```

```python
# Cleaning tweets

import re
import numpy as np

def remove_pattern(text, pattern_regex):
    r = re.findall(pattern_regex, text)
    for i in r:
        text = re.sub(i, '', text)
    
    return text 

# We are keeping cleaned tweets in a new column called 'tidy_tweets'
blizzard_Tweets['Cleaned_tweets'] = np.vectorize(remove_pattern)(blizzard_Tweets[['Tweets']], r'@\w+')
blizzard_Tweets.head(10)
```

```python
blizzard_Tweets['sentiment'] = blizzard_Tweets['Cleaned_tweets'].apply(fetch_sentiment_using_textblob)

blizzard_Tweets.head()
```
